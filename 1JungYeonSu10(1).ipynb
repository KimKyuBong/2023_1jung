```python
# 두 분모가 서로소인 두 유리수의 합, 차를 구하는 수식 생성
import sympy as sp

x, y = sp.symbols("x y")

sp.c2p(sp.mathml(x**2-y**2))
```

```python
from sympy import symbols, latex

# 심볼 정의
x, y = symbols('x y')

# 원래 수식
original_expression = (3/2) + (6/5)

# LaTeX로 수식 출력
latex_expression = latex(original_expression)

print("LaTeX로 표현한 원래 수식:", latex_expression)
```
