```python
# 문자열 알아보기

# 변수는 띄어쓰기 없이 문자를 사용하여 만듭니다.
text1 = "Hello World!" # 변수에 문자열을 집어넣기

print(text1) # text1에 뭐가 들어있는지 출력해보기
print(text1[2]) # 2번째 문자만 가져오기
print(text1[1:4]) # 1번째 이상, 4번째 미만 문자 가져오기
```

```python
# 정수형 알아보기

a = 3 # 변수에 정수 집어넣기
b=2 # 등호나 연산자 사이의 띄어쓰기는 안해도 됨

a+b
```

```python
# 실수형 알아보기

a = 32/100 # 나누기를 이용한 식은 기본적으로 실수형으로 나타남

b = 32/10**15 # 결과를 한번 해석해보자!

print(a, b) # print(a, b, c)처럼 입력하면 같은 줄에 여러 변수값을 나타낼 수 있음.
```

```python
# 리스트 알아보기

list0 = ["홍길동", "남", 35, 4.01] # 이름, 성별, 나이, 교육경력 등 여러자료형을 담을 수 있다
list1 = ["John Doe","남",29,2.10] # 자료별 띄어쓰기는 선택사항이지만, 가독성때문에 띄운다
list2 = ["이마리", "여", 62, 30.12]

AllList=[list0, list1, list2] # 심지어 리스트 안에 리스트를 넣어 이중리스트를 만들수 있다.

print(AllList) # 이런식으로 다차원 배열을 만들 수 있으며
print(AllList[2][3]) # 2번째 자료의 3번째 원소를 가져올 수 있다.
```

```python
# 딕셔너리 알아보기

# 이런식으로 한사람의 정보를 담기도 하고
dict0 = {"이름":"홍길동", "성별":"남", "나이":35, "교육경력":4.01}

# 여러사람의 내용을 아래처럼 입력할 수 도 있다.
dictList = {"이름":["홍길동", "John Doe", "이마리"],
            "성별":["남", "남", "여"],
            "나이":[35, 29, 62],
            "교육경력":[4.01, 2.10, 30.12]}

print(dict0["이름"]) # 이렇게 key값으로 value를 불러올수 있고
print(dictList["성별"][2]) # 2번째 사람의 성별을 이렇게 불러올 수도 있다.
```
